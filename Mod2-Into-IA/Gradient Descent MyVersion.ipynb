```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importar un Dataset
columns = ["sepal length","sepal width","petal length","petal width", "class"]
df = pd.read_csv('../DataSets/iris.data',names = columns)
df.head()
```

```python
# Separar en x & y
x = df['sepal width']
y = df['sepal length']
```

```python
# Hace el y=mx+b
def hyp(m,b,x):
    return m*x + b
```

```python
def gradient_descentt(x,y):
    m_current = 0
    b_current = 0
    learning_rate = 0.03
    n = len(x)

    epochs = 1000

    for i in range(epochs):
        y_pred = hyp(m_current ,b_current, x )
        m_derivative = -(2/n)*sum( (y_pred - y)*x )
        y_derivative = (2/n)*sum(y-y_pred)
        costo = (1/n) * sum([val**2 for val in (y - y_pred)])

        print ("Epoch: ", i, "M VALUE: ", m_current, "B VALUE: ", b_current, "COSTO: ", costo)
        m_current = m_current - (learning_rate * m_derivative)
        b_current= b_current - (learning_rate * y_derivative)
        
```

```python
gradient_descentt(x,y)
```
