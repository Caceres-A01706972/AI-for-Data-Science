```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
%matplotlib inline
import seaborn as sns
sns.set()

sns.pairplot(iris, hue='species', size=1.5);
```

Rather than a histogram, we can get a smooth estimate of the distribution using a kernel density estimation, which Seaborn does with sns.kdeplot

```python

sns.kdeplot(iris['sepal_length'], fill=True)
sns.kdeplot(iris['sepal_width'], fill=True)
sns.kdeplot(iris['petal_length'], fill=True)
sns.kdeplot(iris['petal_width'], fill=True)
```

Histograms and KDE can be combined using distplot

```python
sns.displot(x ='sepal_length',kde=True,bins = 5 ,
            hue = iris['species'] , palette = 'Set3', data=iris)
```

You can also use scatter plots and color them by class easily

```python
sns.scatterplot(x='sepal_length', y ='petal_length' ,data = iris , hue = 'species')
```
